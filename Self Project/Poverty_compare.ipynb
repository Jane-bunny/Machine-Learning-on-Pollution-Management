```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:\\Users\\stoer\\Downloads\\pip_dataset.csv')
```

```python
df = df[['country', 'reporting_level', 'decile1_avg', 'decile2_avg', 'decile3_avg', 'decile4_avg', 'decile5_avg', 'decile6_avg', 'decile7_avg', 'decile8_avg', 'decile9_avg', 'decile10_avg', 'year', 'ppp_version']]
```

```python
df = df.dropna()
```

```python
national_df = df[(df['reporting_level'] == 'national') & (df['ppp_version'] == 2017)]
```

```python
selected_countries = ['Congo', 'United States', 'China', 'Switzerland']
filtered_df = national_df[national_df['country'].isin(selected_countries)] 

poverty_plot=filtered_df.pivot(
    index= 'year',
    columns= 'country',
    values='decile1_avg'
).fillna(0)


poverty_plot.plot(kind= 'line', marker = 'o', linestyle=':', figsize=(15, 8))

plt.xlabel("Data Available Year")
plt.xticks(rotation=90)
plt.ylabel("Daily Expenditure/Income")
plt.title("Poorest (within 1st) in Each Country, Mean Income or Expenditure/Day Comparison ($ Price 2017)")
```

```python
# Define a list of countries you want to plot
selected_countries = ['Congo', 'United States', 'China', 'Switzerland']  # Add the countries you want to plot

# Filter the DataFrame to include only the selected countries
filtered_df = national_df[national_df['country'].isin(selected_countries)]

# Pivot the DataFrame to reshape it
pivot_df = filtered_df.pivot(index='year', columns='country', values=['decile1_avg'])

# Plot the pivoted DataFrame
pivot_df.plot(figsize=(15, 8))
plt.title('Pooest Population in Each Country Over Time by Country')
plt.xlabel('Year')
plt.ylabel('Decile Averages')
plt.legend(title='Country', loc='upper left')
plt.grid(True)
plt.show()
```
