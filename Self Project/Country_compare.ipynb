```python
national_df = df[(df['reporting_level'] == 'national') & (df['ppp_version'] == 2017) & (df['year'] == 2019)]
```

```python
df = national_df.dropna()
```

```python
data = df

top_countries = data.nlargest(3, 'decile10_avg')
bottom_countries = data.nsmallest(3, 'decile1_avg')
top_bottom_countries = pd.concat([top_countries, bottom_countries])

scatter = alt.Chart(data).mark_point().encode(
    alt.X('decile1_avg').scale(zero=False),
    alt.Y('decile10_avg').scale(zero=False, padding=1),
    color='continent',
    size='mean'
)

text = alt.Chart(top_bottom_countries).mark_text(
    align='left', baseline='middle',  # Adjust alignment as needed
    dx=6,  # Offset from the point for better visibility
    fontSize=10,  # Set the font size for the text labels
).encode(
    x='decile1_avg',  # X-coordinate of the text label
    y='decile10_avg',  # Y-coordinate of the text label
    text='country',  # Specify the column for the text labels (country names)
)

chart = (scatter+text).properties(
    width=900,
    height=350, 
    title='Bottom and Top Citizen of Each Country - Daily Mean Income/Expenditure'
).interactive()

chart
```
