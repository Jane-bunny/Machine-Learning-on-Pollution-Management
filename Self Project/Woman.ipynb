```python

# Create a dictionary to map category names to more readable labels
category_labels = {
    'number_nowomen_vdem_owid': 'No Women',
    'number_less10women_vdem_owid': 'Less than 10% Women',
    'number_less20women_vdem_owid': 'Less than 20% Women',
    'number_less30women_vdem_owid': 'Less than 30% Women',
    'number_less40women_vdem_owid': 'Less than 40% Women',
    'number_less50women_vdem_owid': 'Less than 50% Women',
    'number_more50women_vdem_owid': '50% or More Women'
}

categories = ['number_nowomen_vdem_owid','number_less10women_vdem_owid', 'number_less20women_vdem_owid',
       'number_less30women_vdem_owid', 'number_less40women_vdem_owid',
       'number_less50women_vdem_owid', 'number_more50women_vdem_owid']
data = {category: df1[category].values for category in categories}
years = df1['Year'].values
# Assuming you have your 'categories' and 'data' dictionaries ready
# Also, 'years' should be a list of years

# Group the data by decades (summing values for each category within each decade)
decades = [str(year // 10 * 10) for year in years]
unique_decades = sorted(set(decades))
decade_data = {decade: {category: [] for category in categories} for decade in unique_decades}
for year, decade in zip(years, decades):
    for category in categories:
        decade_data[decade][category].append(data[category][year - 1900])

# Calculate the sums for each category within each decade
category_sums = {decade: {category_labels[category]: sum(values) for category, values in data.items()} for decade, data in decade_data.items()}

# Calculate the total sums for each decade
total_decade_sums = {decade: sum(category_sums[decade].values()) for decade in unique_decades}

# Calculate the percentages for each category within each decade
category_percentages = {decade: {category_labels[category]: (category_sums[decade][category_labels[category]] / total_decade_sums[decade]) * 100 for category in categories} for decade in unique_decades}

x = np.arange(len(unique_decades))
width = 0.7

fig, ax = plt.subplots(figsize=(12, 6))

# Initialize the bottom values
bottom = np.zeros(len(unique_decades))

for category in categories:
    # Get the percentage values for the current category
    values = [category_percentages[decade][category_labels[category]] for decade in unique_decades]
    
    # Create the bar chart with percentage values
    ax.bar(x, values, width, label=category_labels[category], bottom=bottom)
    
    # Update the bottom values for stacking
    bottom += values

ax.set_xlabel('Decade')
ax.set_ylabel('Percentage of Shares')
ax.set_title("Women's Participation in Parliament Over Decades (1900-2023)")

# Set x-axis ticks and labels
ax.set_xticks(x)
ax.set_xticklabels(unique_decades, rotation=45)

ax.legend(loc='upper right', bbox_to_anchor=(1.2, 1), title='Categories')

plt.show()
```
